Feature Engineering:

Dropping and Binning! 

```python
df = clean_df.copy()
df = df.drop(columns=["id","yr_renovated"])
```

```python
df["was_renovated"] = (clean_df["yr_renovated"] > 0).astype(int)
```

```python
df["house_age"] = df["year"] - df["yr_built"]
```

```python
df["has_basement"] = (df["sqft_basement"] > 0).astype(int)
```

Grouping Bedrooms: 

```python
df["bedrooms_bin"] = pd.cut(df["bedrooms"],bins=[0, 2, 4, 6, 10, 100],labels=["small", "medium", "large", "very_large", "extreme"])
```

```python
df["bedrooms_bin"] = df["bedrooms_bin"].cat.codes
```

```python
zipcode_price = df.groupby("zipcode")["price"].mean()
```

```python
zipcode_bins = pd.qcut(
    zipcode_price,
    q=3,
    labels=["low_value_area", "mid_value_area", "high_value_area"]
)
```

```python
df["zipcode_group"] = df["zipcode"].map(zipcode_bins)
```

```python
df = df.drop("zipcode", axis=1)
```
